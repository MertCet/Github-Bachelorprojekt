```python
def split_dataframe_by_symbols(df, chunk_size=100, output_prefix='df_final_cleaned'):
    unique_symbols = df['Symbol'].unique()
    
    num_chunks = (len(unique_symbols) + chunk_size - 1) // chunk_size 

    for i in range(num_chunks):
        start_idx = i * chunk_size
        end_idx = min((i + 1) * chunk_size, len(unique_symbols))
        symbol_subset = unique_symbols[start_idx:end_idx]

        subset_df = df[df['Symbol'].isin(symbol_subset)].copy()

        output_filename = f'{output_prefix}_{i+1}.csv'
        subset_df.to_csv(output_filename, index=False)
        print(f'Saved {output_filename} with {len(symbol_subset)} symbols')

    print('All splits processed and saved.')
```

```python
def calculate_momentum(df, intervals, price_col='Adj Close'):
    df = df.copy(deep=True)

    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')  

    df = df.sort_values(by=['Symbol', 'Date'])

    for n in intervals:
        df[f'{n}-month momentum'] = np.nan

    for symbol, group in df.groupby('Symbol'):
        last_days = group.resample('M', on='Date').last().dropna(subset=[price_col])

        for n in intervals:
            momentum_col = f'{n}-month momentum'
            
            num_steps = len(last_days) // n

            for i in range(num_steps):
                idx = i * n

                current_period_end = last_days.index[idx]
                
                if i == 0:
                    last_days.loc[current_period_end, momentum_col] = 0
                    previous_adj_close = last_days.loc[current_period_end, price_col]
                else:
                    current_adj_close = last_days.loc[current_period_end, price_col]
                    momentum_value = current_adj_close - previous_adj_close

                    last_days.loc[current_period_end, momentum_col] = momentum_value

                    previous_adj_close = current_adj_close

            for i in range(num_steps):
                idx = i * n
                current_period_end = last_days.index[idx]
                
                if i == 0:
                    period_start = df['Date'].min()  
                else:
                    period_start = last_days.index[idx - n]
                
                period_end = current_period_end
                momentum_value = last_days.loc[current_period_end, momentum_col]

                mask = (df['Symbol'] == symbol) & (df['Date'] > period_start) & (df['Date'] <= period_end)
                df.loc[mask, momentum_col] = momentum_value

    for n in intervals:
        momentum_col = f'{n}-month momentum'
        df[momentum_col] = df[momentum_col].fillna(method='ffill')

    return df
```

```python
def process_and_save_in_batches(df, intervals, price_col='Adj Close', batch_size=10, output_filename='df_final_cleaned_moment.csv'):
    cumulative_df = pd.DataFrame()

    symbols = df['Symbol'].unique()

    processed_count = 0

    for symbol in symbols:
        symbol_df = df[df['Symbol'] == symbol].copy()

        processed_symbol_df = calculate_momentum(symbol_df, intervals, price_col)

        cumulative_df = pd.concat([cumulative_df, processed_symbol_df], ignore_index=True)

        processed_count += 1

        if processed_count % batch_size == 0:
            cumulative_df.to_csv(output_filename, index=False)
            print(f'Saved {output_filename} with {processed_count} symbols processed')

    if not cumulative_df.empty:
        cumulative_df.to_csv(output_filename, index=False)
        print(f'Saved {output_filename} with {processed_count} symbols processed')

    print(f'Processing complete. Total symbols processed: {processed_count}')
```

```python
def process_split_files(num_files, intervals, price_col='Adj Close', input_prefix='df_final_cleaned', output_prefix='df_final_cleaned_momentum'):
    for i in range(1, num_files + 1):
        input_filename = f'{input_prefix}_{i}.csv'
        df = pd.read_csv(input_filename)

        processed_df = process_and_save_in_batches(df, intervals, price_col, batch_size=10, output_filename=f'{output_prefix}_{i}.csv')

        print(f'Processed and saved {output_prefix}_{i}.csv')

    print('All files processed and saved.')
```

```python
def calculate_annual_btm(df, assets_col='Total assets', liabilities_col='Total liabilities', 
                         price_col='Adj Close', shares_col='Shares Outstanding'):
    df = df.copy(deep=True)
    # Ensure the Date column is in datetime format
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')  

    df = df.sort_values(by=['Symbol', 'Date'])

    # Initialize a new column for btm
    df['Book-to-market'] = np.nan

    for symbol, group in df.groupby('Symbol'):
        # Identify the last day of each year for this symbol
        last_days = group.resample('Y', on='Date').last().dropna(subset=[assets_col, liabilities_col, price_col, shares_col])

        # Calculate the btm for each year's last day
        last_days['Book-to-market'] = (last_days[assets_col] - last_days[liabilities_col]) / (last_days[price_col] * last_days[shares_col])

        for current_year_end in last_days.index:
            year_start = current_year_end.to_period('Y').start_time
            year_end = current_year_end

            btm_value = last_days.loc[current_year_end, 'Book-to-market']

            mask = (df['Symbol'] == symbol) & (df['Date'] >= year_start) & (df['Date'] <= year_end)
            df.loc[mask, 'Book-to-market'] = btm_value

    df['Book-to-market'] = df['Book-to-market'].fillna(method='ffill')

    return df
```

```python
def split_dataframe_by_symbols(df, chunk_size=100, output_prefix='df_final_cleaned_split'):
    unique_symbols = df['Symbol'].unique()
    
    num_chunks = (len(unique_symbols) + chunk_size - 1) // chunk_size  

    for i in range(num_chunks):
        start_idx = i * chunk_size
        end_idx = min((i + 1) * chunk_size, len(unique_symbols))
        symbol_subset = unique_symbols[start_idx:end_idx]

        subset_df = df[df['Symbol'].isin(symbol_subset)].copy()

        output_filename = f'{output_prefix}_{i+1}.csv'
        subset_df.to_csv(output_filename, index=False)
        print(f'Saved {output_filename} with {len(symbol_subset)} symbols')

    print('All splits processed and saved.')
```

```python
def process_btm_for_split_files(num_files, assets_col='Total assets', liabilities_col='Total liabilities',
                                price_col='Adj Close', shares_col='Shares Outstanding', 
                                input_prefix='df_final_cleaned_split', output_prefix='df_final_cleaned_btm'):
    for i in range(1, num_files + 1):
        input_filename = f'{input_prefix}_{i}.csv'
        df = pd.read_csv(input_filename)

        # Calculate the yearly brm
        processed_df = calculate_annual_btm(df, assets_col, liabilities_col, price_col, shares_col)

        output_filename = f'{output_prefix}_{i}.csv'
        processed_df.to_csv(output_filename, index=False)
        print(f'Processed and saved {output_filename}')

    print('All files processed and saved.')
```

```python
def combine_processed_files(num_files, output_prefix='df_final_cleaned_btm', final_output='df_final_cleaned_combined_btm.csv'):
    combined_df = pd.DataFrame()

    for i in range(1, num_files + 1):
        input_filename = f'{output_prefix}_{i}.csv'
        df = pd.read_csv(input_filename)

        combined_df = pd.concat([combined_df, df], ignore_index=True)

    combined_df.to_csv(final_output, index=False)
    print(f'All files combined and saved as {final_output}')
```
